1. Побудувати графіки для тригонометричних функцій, використати різні стилі та маркери

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y1 = np.tan(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y1, linestyle='-', marker='o', markevery=20, label='tan(x)')
ax.plot(x, y2, linestyle='--', marker='s', markevery=20, label='cos(x)')

ax.set_title("Дві лінії")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

2. Побудуйте точкову діаграму, де колір точки залежить від третьої змінної, додайте шкалу кольорів.

```python
np.random.seed(1)
x = np.random.randn(100)
y = 1.2 * x + np.random.randn(100) / 1

c = np.hypot(x, y)  # “інтенсивність” для кольору

fig, ax = plt.subplots(figsize=(6, 3))
sc = ax.scatter(x, y, c=c, s=50, edgecolor='k')
ax.set_title("Scatter: колір за величиною")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, alpha=0.3)
plt.colorbar(sc, ax=ax, label="Колір за теплотою")

plt.show()
```

3. Побудуйте bar chart для 6 категорій і підпишіть значення над стовпчиками.

**Завдання 15.** Побудуйте contour для функції Z(X,Y) на сітці.

**Завдання 16.** Побудуйте 3D-поверхню Z = sin(sqrt(X²+Y²)).

```python
x = np.linspace(-12, 12, 100)
y = np.linspace(-1, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

fig.colorbar(surface, ax=ax, shrink=0.9, label='Z')

ax.set_title('3D-поверхня Z = sin(sqrt(X^2 + Y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

**Завдання 17. **

**Dashboard з GridSpec**: 4 графіки + спільний заголовок

Умова: Створіть одну фігуру-«дашборд» через GridSpec:

зверху широкий графік тренду (line),

знизу зліва гістограма,

знизу справа scatter,

праворуч вузька панель-«індикатор» (bar).

Додайте спільний заголовок фігури, сітку та охайні підписи.

```python
from matplotlib.gridspec import GridSpec

np.random.seed(0)
t = np.linspace(0, 20, 300)
trend = 0.15*t + np.sin(t) + np.random.normal(0, 0.3, len(t))

x = np.random.randn(400)
y = 0.8*x + np.random.randn(400)/2

indicator_values = [3, 11, 4, 7]
indicator_labels = ['A', 'B', 'C', 'D']

fig = plt.figure(figsize=(12,6))
gs = GridSpec(2, 3, figure=fig, width_ratios=[3,3,1], height_ratios=[2,1], wspace=0.4, hspace=0.4)

ax1 = fig.add_subplot(gs[0, :2])
ax1.plot(t, trend, color='tab:blue')
ax1.set_title('Трендова залежність')
ax1.set_xlabel('Час')
ax1.set_ylabel('Значення')
ax1.grid(True)

ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(x, bins=20, color='tab:orange', edgecolor='black')
ax2.set_title('Гістограма X')
ax2.set_xlabel('X')
ax2.set_ylabel('Частота')
ax2.grid(True)

ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y, color='tab:green', alpha=0.6)
ax3.set_title('Scatter X-Y')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.grid(True)

ax4 = fig.add_subplot(gs[:, 2])
ax4.barh(indicator_labels, indicator_values, color='tab:red')
ax4.set_title('Індикатор')
ax4.set_xlabel('Значення')
ax4.grid(True, axis='x')

fig.suptitle('Дашборд: тренд + гістограма + scatter + індикатор', fontsize=16, fontweight='bold')

plt.show()
```

**Завдання 18.**

Зробіть анімацію, де точка «їде» по кривій sin(x).
Потрібно: лінія графіка + рухома точка + текст з координатами точки.

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)
ax.grid(True, alpha=0.3)
ax.set_title('Рух точки по кривій sin(x)', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('sin(x)', fontsize=12)

line, = ax.plot(x, y, 'b-', linewidth=2, label='sin(x)')
point, = ax.plot([], [], 'ro', markersize=10, label='Рухома точка')
coord_text = ax.text(0.02, 0.95, '', transform=ax.transAxes,
                     fontsize=12, verticalalignment='top',
                     bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

ax.legend()

def init():
    point.set_data([], [])
    coord_text.set_text('')
    return point, coord_text

def animate(frame):
    current_x = x[frame]
    current_y = y[frame]
    point.set_data([current_x], [current_y])
    coord_text.set_text(f'x = {current_x:.3f}, y = {current_y:.3f}')
    return point, coord_text

anim = FuncAnimation(fig, animate, init_func=init, frames=len(x), interval=50, blit=True, repeat=True)

plt.close()
HTML(anim.to_html5_video())
```
