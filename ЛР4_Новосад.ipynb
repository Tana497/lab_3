```python
# ФinbTpyєMO лишe за 2015 рiк
df_2015 = df[df['month'].dt.year == 2015]

# Додаємо новий стовпець із загальною кількістю відвідувачів по всіх музеях
df_2015['total_visitors'] = df_2015.iloc[:, 1:].sum(axis=1)

# Знаходимо рядок з мінімальною та максимальною кількістю
min_row = df_2015.loc[df_2015['total_visitors'].idxmin()]
max_row = df_2015.loc[df_2015['total_visitors'].idxmax()]

# Виводимо назви місяців
min_month = min_row['month'].strftime('%B')
max_month = max_row['month'].strftime('%B')

print(f'Найнижча відвідуваність у {min_month}')
print(f'Найвища відвідуваність у {max_month}')
```

```python
# Стиль
sns.set(style="whitegrid", font_scale=1.1)

# Очистка даних
penguins_clean = penguins.dropna()

# Кольори
palette = ["#8e44ad", "#ff6b9d", "#3498db"]

g = sns.PairGrid(penguins_clean, hue="species", corner=True, palette=palette)

# Нижній трикутник
g.map_lower(sns.kdeplot, levels=5, linewidths=1, alpha=0.6)
g.map_lower(sns.scatterplot, marker="+", alpha=0.7)

# Діагональ
g.map_diag(sns.histplot, element="step", linewidth=1, kde=True)

# Легенда
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((0.65, 0.6))

# Заголовок
g.fig.suptitle("Розподіл та взаємозв'язки вимірювань пінгвінів", y=1.02)

plt.show()
```
