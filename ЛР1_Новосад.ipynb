Обернена матриця

```python
a = np.matrix([[1, -5], [9, 4]])
a_inv = np.linalg.inv(a)
a_inv
```

Ранг матриці

```python
rank = np.linalg.matrix_rank(a)
rank
```

```python
a  = np.array([[0, 4, 6], [-3, -11, -1], [11, 12, 13]])
eigenvalues, eigenvectors = np.linalg.eig(a)

print("\n")
print(eigenvalues)
print("\n")
print(eigenvectors)
```
