```python
# Вибираємо числові колонки
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    if df[col].isnull().any():
        # Замінюємо тип на float і заповнюємо середнім за стовпцем
        df[col] = df[col].astype(float)
        df[col] = df[col].fillna(df[col].mean())
```

```python
# 5. Ще раз перевірте кількість пропусків
print("Кількість пропусків після обробки:")
print(df[numeric_cols].isnull().sum())
```

```python
unique_countries = df['Country'].nunique()
print(f"У колонці 'Country' представлено {unique_countries} унікальних країн.")
```

```python
plt.figure(figsize=(12, 6))
df['video views'].plot(kind='hist', bins=50, color='teal', edgecolor='white')
plt.title('Розподіл переглядів відео')
plt.xlabel('Кількість переглядів')
plt.ylabel('Кількість каналів')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
max_v = df['video views'].max()
min_v = df['video views'].min()
avg_v = df['video views'].mean()

print(f"Максимальна кількість переглядів: {max_v}")
print(f"Мінімальна кількість переглядів: {min_v}")
print(f"Середня кількість переглядів: {avg_v:.2f}")
```

```python
# Групуємо по країні та знаходимо суму завантажень
top_country = df.groupby('Country')['uploads'].sum().idxmax()
print(f"Країна, де найбільше відео було завантажено на YouTube: {top_country}")
```

```python
# Відео з найбільшою кількістю завантажень
max_uploads_video = df.loc[df['uploads'].idxmax(), 'Title']

# Відео з найменшою кількістю завантажень
min_uploads_video = df.loc[df['uploads'].idxmin(), 'Title']

print(f"Відео (канал) з найбільшою кількістю завантажень: {max_uploads_video}")
print(f"Відео (канал) з найменшою кількістю завантажень: {min_uploads_video}")
```

```python
# Перейменування: нижній регістр та заміна пробілів на _
df.columns = [col.lower().replace(' ', '_') for col in df.columns]

# Перевіримо нові назви
print("Нові назви стовпців:")
print(df.columns)
```

```python
sns.set_style("whitegrid")

# 1. Гістограма віку (age)
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Розподіл віку клієнтів')
plt.show()
```

```python
# 2. Порівняння розподілу покупок (item_purchased)
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='item_purchased', order=df['item_purchased'].value_counts().index, palette='viridis')
plt.title('Найпопулярніші товари')
plt.show()
```

```python
# 3. Залежність між віком та сумою покупок (purchase_amount)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='purchase_amount_(usd)', alpha=0.6)
plt.title('Залежність суми покупок від віку')
plt.show()
```

```python
# 4. Графік розподілу оцінок товарів (review_rating)
plt.figure(figsize=(10, 5))
sns.kdeplot(df['review_rating'], fill=True, color='orange')
plt.title('Розподіл оцінок товарів')
plt.show()
```

```python
# 5. Категоріальна діаграма для статі (gender)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gender', palette='pastel')
plt.title('Розподіл покупок за статтю')
plt.show()
```

```python
# 6. Залежність між рейтингом і сумою покупок
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='review_rating', y='purchase_amount_(usd)')
plt.title('Залежність суми покупок від рейтингу')
plt.show()
```

```python
import altair as alt
from vega_datasets import data
```

```python
# Завантажуємо дані
wheat = data.wheat()
wheat
```

```python
bars = alt.Chart(wheat).mark_bar().encode(
    x='wheat:Q',
    y='year:O'
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    text='wheat:Q'
)

chart1 = (bars + text).properties(title="Врожайність пшениці за роками")
chart1
```

```python
# Завантажуємо дані
a = data.iowa_electricity()
a
```

```python
chart2 = alt.Chart(a).mark_bar(opacity=0.7).encode(
    x='year:T',
    y='net_generation:Q',
    color='source:N'
).properties(
    title="Виробництво електроенергії в Айові за джерелами",
    width=600
)

chart2
```
